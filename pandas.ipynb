```python
#informacoes q01 e q02 dicionario
dicionario01e02 = {'A':'Nunca Estudou',
                   'B':'não completou a 4/5 serie do fundamental',
                   'C':'completou a 4/5 do fundamental masn a 8/9',
                   'D':'completou a 8/9 mas n o ensino medio',
                   'E':'completou o ensino medio mas n a facul',
                   'F':'completou a faculdade mas n a pos',
                   'G':'Completou pos graduacao',
                   'H':'N sei',}
```

```python
dadosSelecionados['NO_Q001'] = [dicionario01e02[resp] for resp in dadosSelecionados.Q001]
dadosSelecionados['NO_Q002'] = [dicionario01e02[resp] for resp in dadosSelecionados.Q002]
dadosSelecionados.head()
```

```python
#sort_values(a tabela que vc quer por em ordem) ---- ascending (menor p maior) ascending False (maior pra menor)
#pais
dadosSelecionados.filter(items=['NU_INSCRICAO','NO_Q001']).groupby('NO_Q001').count().sort_values(by='NU_INSCRICAO', ascending=False)
```

```python
#maes
dadosSelecionados.filter(items=['NU_INSCRICAO','NO_Q002']).groupby('NO_Q002').count().sort_values(by='NU_INSCRICAO', ascending=False)
```

#nota redacao por escolaridade dos pais, em cada estado

```python
enem.columns.values
```

```python
colunasSelecionadas = ['NU_INSCRICAO','SG_UF_ESC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO', 'Q001', 'Q002']

dadosSelecionados = enem.filter(items=colunasSelecionadas).dropna()
dadosSelecionados
```

```python
#informacoes q01 e q02 dicionario
dicionario01e02 = {'A':'Nunca Estudou',
                   'B':'não completou a 4/5 serie do fundamental',
                   'C':'completou a 4/5 do fundamental masn a 8/9',
                   'D':'completou a 8/9 mas n o ensino medio',
                   'E':'completou o ensino medio mas n a facul',
                   'F':'completou a faculdade mas n a pos',
                   'G':'Completou pos graduacao',
                   'H':'N sei',}
```

```python
dadosSelecionados['NO_Q001'] = [dicionario01e02[e] for e in dadosSelecionados.Q001]
```

```python
dadosSelecionados['NO_Q002'] = [dicionario01e02[e] for e in dadosSelecionados.Q002]
```

```python
#WHERE UF == RJ
dadosSelecionados.filter(items=['NO_Q002', 'NU_NOTA_REDACAO'])\
.where(dadosSelecionados.SG_UF_ESC == 'RJ')\
.groupby('NO_Q002')\
.mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)
```

```python
#WHERE UF == MG
dadosSelecionados.filter(items=['NO_Q002', 'NU_NOTA_REDACAO'])\
.where(dadosSelecionados.SG_UF_ESC == 'MG')\
.groupby('NO_Q002')\
.mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)
```

```python
#TODAS AS REGIOES
dadosSelecionados.filter(items=['SG_UF_ESC','NO_Q002', 'NU_NOTA_REDACAO'])\
.groupby(['SG_UF_ESC','NO_Q002'])\
.mean()
```

```python
#Pacote para facilitar a visualizacao
import matplotlib.pyplot as plt
```

```python


ax= dadosSelecionados.filter(items=['SG_UF_ESC','Q002', 'NU_NOTA_REDACAO'])\
.where(dadosSelecionados.Q002 != 'H')\
.groupby(['Q002','SG_UF_ESC'])\
.mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)\
.unstack().plot(figsize=(16,8))

ax.set_title("Nota redacao x escolaridade x estado")
#ax.legend(bbox_to_anchor=(1,1), loc=0, borderaxespad=0)

handles, labels = ax.get_legend_handles_labels()

import re
edited_labels = [re.search(',\s(.+?)\)', label).group(1) for label in labels]
ax.legend(edited_labels, bbox_to_anchor=(1,1), loc=0, borderaxespad=0)

textdict = ""
for k, v in dicionario01e02.items():
  textdict = textdict + "{key} : {value}\n".format(key=k, value=v)

ax.text(0,1, textdict, transform=ax.transAxes, fontsize=14)
```
